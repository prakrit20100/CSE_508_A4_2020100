```python
from transformers import Trainer, TrainingArguments, GPT2LMHeadModel, GPT2Tokenizer, DataCollatorForLanguageModeling
from torch.utils.data import Dataset
import torch
import os
class TextDatasetFromDataFrame(Dataset):
    def __init__(self, dataframe, tokenizer, block_size=512):
        self.input_ids = []
        self.attention_masks = []
        for _, row in dataframe.iterrows():
            combined_text = row['input_text'] + "<|startoftext|>" + row['target_text']
            text = tokenizer.encode_plus(
                combined_text,
                max_length=block_size,
                truncation=True,
                padding='max_length'
            )
            self.input_ids.append(torch.tensor(text['input_ids']))
            self.attention_masks.append(torch.tensor(text['attention_mask']))


    def __len__(self):
        return len(self.input_ids)

    def __getitem__(self, idx):
        return {
            'input_ids': self.input_ids[idx],
            'attention_mask': self.attention_masks[idx]
        }

def train(train_df, model_name, output_dir, overwrite_output_dir, per_device_train_batch_size, num_train_epochs, save_steps, learning_rate):
    tokenizer = GPT2Tokenizer.from_pretrained(model_name)
    tokenizer.pad_token = tokenizer.eos_token

    train_dataset = TextDatasetFromDataFrame(train_df, tokenizer, block_size=512)
    data_collator = DataCollatorForLanguageModeling(tokenizer=tokenizer, mlm=False)

    model = GPT2LMHeadModel.from_pretrained(model_name).to('cuda')

    training_args = TrainingArguments(
        output_dir=output_dir,
        overwrite_output_dir=overwrite_output_dir,
        per_device_train_batch_size=per_device_train_batch_size,
        num_train_epochs=num_train_epochs,
        save_steps=save_steps,
        logging_dir=f"{output_dir}/logs",
        learning_rate=learning_rate,
        evaluation_strategy="no",
        load_best_model_at_end=False
    )


    trainer = Trainer(
        model=model,
        args=training_args,
        data_collator=data_collator,
        train_dataset=train_dataset
    )

    trainer.train()
    trainer.save_model(os.path.join(output_dir))
    tokenizer.save_pretrained(output_dir)

model_name = 'gpt2'
output_dir = "/content/drive/MyDrive/IR_Assignment_4/learning_rate_changed_model"
save_steps = 500
overwrite_output_dir = True
per_device_train_batch_size = 8
num_train_epochs = 3
learning_rate = 2e-4

train(train_df, model_name, output_dir, overwrite_output_dir, per_device_train_batch_size, num_train_epochs, save_steps, learning_rate)
```

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
from datasets import load_metric
import torch
import pandas as pd

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

def load_model_and_tokenizer(model_path):
    model = GPT2LMHeadModel.from_pretrained(model_path).to(device)
    tokenizer = GPT2Tokenizer.from_pretrained(model_path)
    return model, tokenizer

def generate_text(model, tokenizer, input_sequence, max_length):
    model.eval()
    input_ids = tokenizer.encode(input_sequence + "<|startoftext|>", return_tensors='pt').to(device)

    outputs = model.generate(
        input_ids,
        max_length=max_length + len(input_ids[0]),
        pad_token_id=tokenizer.eos_token_id,
        eos_token_id=tokenizer.eos_token_id,
        do_sample=True,
        top_k=50,
        top_p=0.95
    )

    generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    # print(generated_text)
    generated_parts = generated_text.split("<|startoftext|>")
    return generated_parts[-1].strip()




def evaluate_summaries(df, model, tokenizer, max_length):
    rouge = load_metric("rouge")
    summaries = []
    references = df['target_text'].tolist()

    for text in df['input_text'].tolist():
        summaries.append(generate_text(model, tokenizer, text, max_length))

    results = rouge.compute(predictions=summaries, references=references)
    for key, value in results.items():
        print(f"{key}: Precision: {value.mid.precision:.2f}, Recall: {value.mid.recall:.2f}, F1-Score: {value.mid.fmeasure:.2f}")

model_path = "/content/drive/MyDrive/IR_Assignment_4/learning_rate_changed_model"

model, tokenizer = load_model_and_tokenizer(model_path)

print(test_df["input_text"].iloc[2])
print(test_df["target_text"].iloc[2])
print(generate_text(model, tokenizer, test_df["input_text"].iloc[2], max_length=5))
evaluate_summaries(test_df, model, tokenizer, max_length=5)
```
